```python
def funAk(k):
    A=np.array([[1,1],[1,0]])
    autovalores=eig(A)[0]
    D=np.eye(2,2)*autovalores
    P=eig(A)[1]
    Pinv=inv(P)
    Dk=D**k
    Ak=np.dot(np.dot(P,Dk),Pinv)
    return Ak
```

```python
for i in range(1,15):
    Ak=funAk(i)
    F=np.array([0,1])
    v=np.dot(Ak,F)
    print('T√©rmino ', i, '-√©simo: ',v[0])
```

```python
def funPol(coef):
    import functools 
    x=['','x','x^2','x^3','x^4']
    polList=list(map(lambda a,x:str(a)+x if a<0 else '+'+str(a)+x ,coef,x))
    pol=functools.reduce(lambda a,b:a+b,polList)
    return pol
```
